```python
from IPython.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import sympy as sym
sym.init_printing(use_unicode=True)
from time import sleep
from tqdm import tqdm
```

```python
N = 1000000
a = 0
b = np.pi
x = np.random.uniform(a,b,N) #N es el número de muestras aleatorias que tomará
```

```python
def func_integrate(x):
    return np.exp(-x)*np.sin(x)
```

```python
fi = func_integrate(x)
```

```python
I = (b-a)*sum(fi)/N
print(I)
```

```python
N_array=np.arange(1,N,1000)
X=np.zeros(len(range(1,N,1000)))
I_raizN=np.zeros(len(range(1,N,1000)))
Iteo = 0.5*(1+np.exp(-np.pi))
l=0
for i in tqdm(N_array):
    o = np.random.uniform(a,b,i)
    I_N=(b-a)*sum(func_integrate(o))/i
    error=np.abs(1-I_N/Iteo)
    X[l]=error
    I_raizN[l]=1/(i**0.5)
    l+=1
    

plt.plot(N_array,X)
plt.plot(N_array,I_raizN)
```
